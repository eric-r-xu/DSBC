classification algorithms:

```python
import pandas as pd
df=pd.read_csv('house.txt')

df.columns = pd.Index([u'A1', u'A2', u'A3', u'A4', u'A5', u'A6', u'A7', u'A8', u'A9', u'A10', u'A11', u'A12', u'A13', u'A14', u'A15', u'A16', u'A17'])
df=df.replace('y', 1)
df=df.replace('n', 0)
import numpy as np
df=df.replace('?',np.nan)
df=df.replace('democrat',0)
df=df.replace('republican',1)
df.mean()
df=df.fillna(df.mean())



from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.learning_curve import learning_curve
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.cross_validation import train_test_split


def RepublicanPred(XDATA):
    OUTPUT=[]
    for each in XDATA:
        OUTPUT.append(1)
    return OUTPUT

def DemocratPred(XDATA):
    OUTPUT=[]
    for each in XDATA:
        OUTPUT.append(0)
    return OUTPUT

#preparing training and test sets
dftrain, dftest= train_test_split(df, test_size=0.25)
dfx_train=dftrain[:,1:18]
dfy_train=dftrain[:,0]
dfx_test=dftest[:,1:18]
dfy_test=dftest[:,0]
dfx_all=df.values[:,1:18]
dfy_all=df.values[:,0]

#RepublicanPred DemocratPred 
y_pred = RepublicanPred(dfx_test)
print "Republican: accuracy score", accuracy_score(dfy_test, y_pred)
print "Republican: precision score",precision_score(dfy_test, y_pred)
print "Republican: recall score", recall_score(dfy_test, y_pred)
print "Republican: f1 score", f1_score(dfy_test, y_pred)
y_pred = DemocratPred(dfx_test)
print "-------------------------------------"
print "Democrat: accuracy score",accuracy_score(dfy_test, y_pred)
print "Democrat: precision score",precision_score(dfy_test, y_pred,pos_label=0)
print "Democrat: recall score",recall_score(dfy_test, y_pred,pos_label=0)
print "Democrat: f1 score",f1_score(dfy_test, y_pred,pos_label=0)
print "-------------------------------------"

for x in range(1, 21):
    clf = KNeighborsClassifier(n_neighbors=x)
    clf.fit(dfx_train, dfy_train)
    y_pred = clf.predict(dfx_test)
    print "KNN accuracy score with k =",x, " ",accuracy_score(dfy_test, y_pred)
    print "KNN precision score with k =",x, " ",precision_score(dfy_test, y_pred,pos_label=None)
    print "KNN recall score with k =",x, " ",recall_score(dfy_test, y_pred,pos_label=None)
    print "KNN f1 score with k =",x, " ",f1_score(dfy_test, y_pred,pos_label=None)
    

from sklearn.linear_model import LogisticRegression 

logitmodel = LogisticRegression()
logitmodel.fit(dfx_train, dfy_train)
y_pred = logitmodel.predict(dfx_test)
print "-------------------------------------"
print "Log: accuracy score",accuracy_score(dfy_test, y_pred)
print "Log: precision score",precision_score(dfy_test, y_pred,pos_label=None)
print "Log: recall score",recall_score(dfy_test, y_pred,pos_label=None)
print "Log: f1 score",f1_score(dfy_test, y_pred,pos_label=None)

GNB = GaussianNB()
GNB.fit(dfx_train, dfy_train)
y_pred = GNB.predict(dfx_test)
print "-------------------------------------"
print "GaussianNB: accuracy score",accuracy_score(dfy_test, y_pred)
print "GaussianNB: precision score",precision_score(dfy_test, y_pred,pos_label=None)
print "GaussianNB: recall score",recall_score(dfy_test, y_pred,pos_label=None)
print "GaussianNB: f1 score",f1_score(dfy_test, y_pred,pos_label=None)
model=SVC()
model.fit(dfx_train, dfy_train)
y_pred = model.predict(dfx_test)
print "-------------------------------------"
print "SVC: accuracy score",accuracy_score(dfy_test, y_pred)
print "SVC: precision score",precision_score(dfy_test, y_pred,pos_label=None)
print "SVC: recall score",recall_score(dfy_test, y_pred,pos_label=None)
print "SVC: f1 score",f1_score(dfy_test, y_pred,pos_label=None)
DTC = DecisionTreeClassifier()
DTC.fit(dfx_train, dfy_train)
y_pred = DTC.predict(dfx_test)
print "-------------------------------------"
print "DTC: accuracy score",accuracy_score(dfy_test, y_pred)
print "DTC: precision score",precision_score(dfy_test, y_pred,pos_label=None)
print "DTC: recall score",recall_score(dfy_test, y_pred,pos_label=None)
print "DTC: f1 score",f1_score(dfy_test, y_pred,pos_label=None)
RFC = RandomForestClassifier()
RFC.fit(dfx_train, dfy_train)
y_pred = RFC.predict(dfx_test)
print "-------------------------------------"
print "RFC: accuracy score",accuracy_score(dfy_test, y_pred)
print "RFC: precision score",precision_score(dfy_test, y_pred,pos_label=None)
print "RFC: recall score",recall_score(dfy_test, y_pred,pos_label=None)
print "RFC: f1 score",f1_score(dfy_test, y_pred,pos_label=None)
```

```python
%matplotlib inline
from datetime import datetime
import pickle
movie_data=[]

movie_data = pd.read_csv('mv_df_budget.csv')

movie_data.head()
movie_data.groupby('month')['title'].count().plot(kind='bar')
pl.ylabel('number of movies')


X = movie_data[['runtime','profit','budget','theater_number','months_release','year']]
Y = movie_data['month']

for k in range(1,31):
    model = KNeighborsClassifier(n_neighbors=k)
    accuracy = np.mean(cross_val_score(model, X,Y, cv=10))
    precision = np.mean(cross_val_score(model, X,Y, cv=10, scoring='precision'))
    recall = np.mean(cross_val_score(model, X,Y, cv=10, scoring='recall'))
    print 'KNN(k=%i) accuracy: %.3f' % (k,accuracy)
    print 'KNN(k=%i) precision: %.3f' % (k,precision)
    print 'KNN(k=%i) recall: %.3f' % (k,recall)
    
model = LogisticRegression()
accuracy = np.mean(cross_val_score(model, X,Y, cv=10))
precision = np.mean(cross_val_score(model, X,Y, cv=10, scoring='precision'))
recall = np.mean(cross_val_score(model, X,Y, cv=10, scoring='recall'))
print 'Logistic Regression (L2 penalty) accuracy: %.3f' % accuracy
print 'Logistic Regression (L2 penalty) precision: %.3f' % precision
print 'Logistic Regression (L2 penalty) recall: %.3f' % recall

model = LogisticRegression(penalty='l1')
accuracy = np.mean(cross_val_score(model, X,Y, cv=10))
precision = np.mean(cross_val_score(model, X,Y, cv=10, scoring='precision'))
recall = np.mean(cross_val_score(model, X,Y, cv=10, scoring='recall'))
print 'Logistic Regression (L1 penalty) accuracy: %.3f' % accuracy
print 'Logistic Regression (L1 penalty) precision: %.3f' % precision
print 'Logistic Regression (L1 penalty) recall: %.3f' % recall
# Ok, let's also check different C values. C is just like alpha in Ridge Regression,
# It let's us fine tune how much regularization (how much shrinking) we will do on
# coeefficients. Default C is 1.0. Small C means more shrinking, tends to underfit,
# Large C means less shrinking, tends to overfit.
for C in np.linspace(0.05,100,10):
    model = LogisticRegression(penalty='l1', C=C)
    accuracy = np.mean(cross_val_score(model, X,Y, cv=10))
    precision = np.mean(cross_val_score(model, X,Y, cv=10, scoring='precision'))
    recall = np.mean(cross_val_score(model, X,Y, cv=10, scoring='recall'))
    print 'Logistic Regression (L1 penalty, C=%.2f) accuracy: %.3f' % (C,accuracy)
    print 'Logistic Regression (L1 penalty, C=%.2f) precision: %.3f' % (C,precision)
    print 'Logistic Regression (L1 penalty, C=%.2f) recall: %.3f' % (C,recall)
```

```python
def baseline_model(X, same_guess="Oct"):
    pred = np.array([same_guess] * len(X))
    return pred

class Baseline(object):
    
    def __init__(self, guess):
        self.guess = guess
        
    def fit(self, X, Y):
        # Do nothing. This classifier
        # doesn't really train on data,
        # it will only guess the same thing!
        pass
    
    def predict(self, X):
        # create an array full of the exact same
        # guess and return it!
        pred = np.array([self.guess] * len(X))
        return pred        
    
    def score(self, X,Y):
        # Calculate the accuracy of your
        # stupid predictions against the
        # actual values
        Y_pred = self.predict(X)
        Y_true = Y
        return accuracy_score(Y_true, Y_pred)
    
    def get_params(self, deep=False):
        # Scikitlearn models return a dictionary
        # of {parameter: value} with this function
        # We only have one parameter, out guess.
        # You don't even need to worry about the 
        # "deep" kw argument, we're not using it.
        return {'guess': self.guess}
    
# Ok, now that our class has fit, predict, score and get_params
# just as cross_val_score expects, we can run it!
model = Baseline(guess="Oct")
accuracy = np.mean(cross_val_score(model, X,Y, cv=10))
precision = np.mean(cross_val_score(model, X,Y, cv=10, scoring='precision'))
recall = np.mean(cross_val_score(model, X,Y, cv=10, scoring='recall'))
print 'Baseline Oct guess accuracy: %f' % accuracy
print 'Baseline Oct guess precision: %f' % precision
print 'Baseline Oct guess recall: %f' % recall
print 'KNN and Logistic Regression do around ~2 percent better than Baseline model'
```
